```python
import FinanceDataReader as fdr
import pandas as pd
import numpy as np
import requests
from tqdm import tqdm
import time

from random import uniform, seed
seed(1)
```

```python
# 한국거래소 상장종목 전체
df_1 = fdr.StockListing('NASDAQ')
df_2= fdr.StockListing('NYSE')
df_3= fdr.StockListing('AMEX')
df_4= fdr.StockListing('S&P500')
```

```python
df = pd.concat([df_3, df_4])
df_3.shape, df_4.shape, df.shape
```

```python
df_s = df.Symbol.reset_index().drop('index', axis=1)
df_s=df_s.Symbol
```

```python
isin_series = pd.Series()
for ticker in tqdm(df_s):
    url= 'https://api-prod.ii.co.uk/api/1/faceted-instrument-search-results'
    data={
        "query": f"{ticker}"
    }
    header ={
    'accept': 'application/json;charset=UTF-8',
    'accept-encoding': 'gzip, deflate, br',
    'accept-language': 'ko-KR,ko;q=0.9,en-US;q=0.8,en;q=0.7',
    'ii-consumer-type': 'web.public',
    'origin': 'https://www.ii.co.uk',
    'referer': 'https://www.ii.co.uk/shares/apple-inc/NASDAQ:AAPL',
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/104.0.0.0 Safari/537.36'
    }
    response = requests.get(url,headers=header,params=data)
    results = response.json()['company']['results']
    if results != []:
        for i in results:
            isin = i['isin']
            symbol = i['symbol']
            if (ticker == symbol) and (isin[:2] == 'US'):
                isin_series[f'{ticker}'] = isin
                print(ticker, isin)
                break
            else:
                isin_series[f'{ticker}'] = 'mismatch'
                print(ticker, "mismatch symbol")
    else:
        isin_series[f'{ticker}'] = 'noresult'
        print(ticker, "no results")
```

```python
isin_34 = pd.DataFrame(isin_series).reset_index()
isin_34.columns = ['ticker','isin']

isin_34.to_csv('data/isin/isin_34.csv')
```
